***Змінні та типи даних***

---


Змінні та типи даних є фундаментальними концепціями у більшості мов програмування, включаючи Python. Вони використовуються для збереження та обробки інформації в програмах. Ось детальніше про ці поняття:

**Змінні**:
- Змінна - це іменований контейнер, який зберігає дані або значення. В Python, ви можете створити змінну, наділивши її ім'ям і присвоївши їй значення.



**Типи даних:**

- Тип даних визначає природу даних, які можна зберігати у змінних. В Python існують різні типи даних. Основні типи даних включають:
    1. **int (цілі числа)**: Використовуються для зберігання цілих чисел, наприклад 1, -5, 1000.
    2. **float (числа з рухомою комою)**: Використовуються для зберігання чисел з рухомою комою, наприклад 3.14, -0.5, 2.0.
    3. **str (рядки)**: Використовуються для зберігання текстової інформації, наприклад "Hello, World!".
    4. **bool (булеві значення)**: Використовуються для представлення істинності або неправдивості, True або False.
    5. **list (списки)**: Використовуються для створення змінних, які можуть містити кілька значень, наприклад [1, 2, 3].
    6. **tuple (кортежі)**: Схожі на списки, але не можуть бути змінені після створення, наприклад (1, 2, 3).
    7. **dict (словники)**: Використовуються для створення пар ключ-значення, наприклад {"ім'я": "John", "вік": 30}.
    8. **set (множини)**: Використовуються для зберігання неповторюваних значень, наприклад {1, 2, 3}.

Python є динамічно типізованою мовою, що означає, що ви можете змінювати тип змінної в процесі виконання програми. Наприклад, змінна може спочатку містити ціле число, а потім бути переведена в рядок.

***Операції***

Операції в мові програмування Python використовуються для виконання різних дій зі змінними та значеннями. Ви можете використовувати операції для обчислень, порівнянь, об'єднання даних та інших завдань.

1. **Арифметичні операції**:
   - `+` (додавання): Додає два числа разом.
   - `-` (віднімання): Віднімає одне число від іншого.
   - `*` (множення): Перемножує два числа.
   - `/` (ділення): Ділить перше число на друге, повертає десятковий результат.
   - `//` (цілочисельне ділення): Ділить перше число на друге, і повертає ціле число.
   - `%` (модуль): Повертає залишок від ділення першого числа на друге.
   - `**` (піднесення до ступеня): Підносить перше число до ступеня другого.

Приклад:



```python
a = 10
b = 3

add = a + b  # Результат: 13
subtraction = a - b  # Результат: 7
multi = a * b  # Результат: 30
division = a / b  # Результат: 3.3333333333333335
integerPart = a // b  # Результат: 3
remainder = a % b  # Результат: 1
exponentiation = a ** b  # Результат: 1000
```

2. **Операції порівняння**:
   - `==` (рівність): Перевіряє, чи два значення рівні.
   - `!=` (нерівність): Перевіряє, чи два значення не рівні.
   - `<` (менше): Перевіряє, чи перше значення менше другого.
   - `>` (більше): Перевіряє, чи перше значення більше другого.
   - `<=` (менше або рівне): Перевіряє, чи перше значення менше або рівне другому.
   - `>=` (більше або рівне): Перевіряє, чи перше значення більше або рівне другому.

Приклад:


```python
x = 5
y = 10
equality = x == y  # Результат: False
notequality = x != y  # Результат: True
less = x < y  # Результат: True
more = x > y  # Результат: False
lessnequal = x <= y  # Результат: True
morenequal = x >= y  # Результат: False
```

3. **Логічні операції**:
   - `and` (логічне "І"): Повертає True, якщо обидва вирази є True.
   - `or` (логічне "АБО"): Повертає True, якщо хоча б один з виразів є True.
   - `not` (логічне "НЕ"): Інвертує значення, True стає False і навпаки.

Приклад:

```python
a = True
b = False
logicaland = a and b  # Результат: False
logicalor = a or b  # Результат: True
logicalnot = not a  # Результат: False
```

***Умовні конструкції***

Умовні конструкції в мові програмування Python дозволяють виконувати різні блоки коду в залежності від умов, які визначені у вашому програмі. Python має декілька видів умовних конструкцій, які допомагають контролювати логіку виконання програми.Умовні конструкції допомагають вам керувати логікою вашого коду, роблять його більш гнучким та адаптованим до різних сценаріїв використання. Умовні конструкції є важливою частиною програмування в Python і в інших мовах програмування, і вони допомагають створювати більш складні програми з різними умовами та логічними операціями.


1. **if-elif-else**: Ця конструкція дозволяє виконати блок коду, якщо задана умова є True, інакше можна перевірити інші умови за допомогою `elif` (якщо не одна з них True) і, в кінці, виконати альтернативний блок коду з `else` (якщо жодна з умов не виконується).

Приклад:



```python
x = 10
if x > 10:
    print("x більше 10")
elif x == 10:
    print("x дорівнює 10")
else:
    print("x менше 10")
```

2. **Цикли**:
   - **for**: Цикл `for` використовується для ітерації по послідовностях, таких як рядки, списки, словники тощо.

Приклад:

```python
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    square = num ** 2
    print(f"{num} в квадраті дорівнює {square}")


```

 - **while**: Цикл `while` виконується, поки задана умова є True.

Приклад:

```python
score = 0
while score < 5:
    print(score)
    score += 1
```

3. **Повернення значень**: Функції в Python можуть повертати значення за допомогою оператора `return`. Значення, яке повертається, може бути використане в іншій частині вашої програми.

Приклад:

```python
def double(x):
    result = x * 2
    return result
```

4. **Повернення значень за замовчуванням**: Ви можете задати значення за замовчуванням для аргументів функції, які будуть використовуватися, якщо значення не передається при виклику функції.

Приклад:


```python
def welcome(name="Іван", language="англійська"):
    if language == "англійська":
        return f"Hello, {name}!"

    elif language == "французька":
        return f"Bonjour, {name}!"

    elif language == "іспанська":
        return f"Hola, {name}!"

    else:
        return f"Привіт, {name}!"

message1 = welcome()  # Виклик функції без вказання мови та імені
message2 = welcome("Іван")  # Виклик функції з ім'ям, але без вказання мови
message3 = welcome("Іван", "французька")  # Виклик функції з ім'ям та вказанням мови

print(message1)  # Результат: "Hello, Гість!"
print(message2)  # Результат: "Hello, Іван!"
print(message3)  # Результат: "Bonjour, Іван!"
```

# Ось програма написана з допомогою цих основ Phyton

```python
# Функція для додавання
def add(a, b):
    return a + b

# Функція для віднімання
def sub(a, b):
    return a - b

# Функція для множення
def multi(a, b):
    return a * b

# Функція для ділення
def div(a, b):
    if b == 0:
        return "Помилка: ділення на нуль"
    return a / b

# Головна функція калькулятора
def calculator():
    operation = input("Виберіть операцію (+, -, *, /): ")
    number1 = float(input("Введіть перше число: "))
    number2 = float(input("Введіть друге число: "))

    if operation == "+":
        result = add(number1, number2)
    elif operation == "-":
         result = sub(number1, number2)
    elif operation == "*":
         result = multi(number1, number2)
    elif operation == "/":
         result = div(number1, number2)
    else:
         result = "Невідома операція"

    print(f"Результат: {result}")

# Виклик головної функції калькулятора
calculator()
```
